## Dummy variables

```python
def dummy_feature(df, variables):
    dummies = pd.DataFrame()
    for var in variables:
        dummy = pd.get_dummies(df[var], drop_first=True)
        dummies = pd.concat([dummies, dummy], axis=1)
        df = df.drop(var, axis=1)
    df = pd.concat([df, dummies], axis=1)
    return df
```

# Rank Ensemble

```python
def rank_ensemble(probs):
    en_probs = probs.copy()
    for col in probs.columns:
        en_probs = en_probs.sort_values(by=col)
        en_probs.loc[:,col] = np.arange(1,en_probs.shape[0]+1)
    en_probs = en_probs.sort_index()
    en_probs = probs.apply(lambda x: np.mean(x), axis=1)
    max_ = max(en_probs)
    min_ = min(en_probs)
    en_probs = en_probs.apply(lambda x: (x-min_)/(max_-min_))
    return en_probs
```

```python
en_rank = rank_ensemble(probs)
en_rank_xgb_lr = rank_ensemble(probs[['LR','XGB']])
en_rank_lgb_lr = rank_ensemble(probs[['LR','LGB']])
en_rank_rf_lr = rank_ensemble(probs[['LR','RF']])
en_rank_rf_xgb = rank_ensemble(probs[['XGB','RF']])
en_rank_rf_lgb = rank_ensemble(probs[['LGB','RF']])
en_rank_xgb_lgb = rank_ensemble(probs[['LGB','XGB']])
en_rank_xgb_lgb_lr = rank_ensemble(probs[['LGB','XGB','LR']])
en_rank_xgb_rf_lr = rank_ensemble(probs[['RF','XGB','LR']])
en_rank_rf_lgb_lr = rank_ensemble(probs[['LGB','RF','LR']])
en_rank_xgb_lgb_rf = rank_ensemble(probs[['LGB','XGB','RF']])
```
